 ## <span style='background :yellow' > Question 1: What percentage of trades are with 30+ DPD (more than 30 days past due) among all the trades available? </span>

```python
# Importing libraries 
import os
import json
from bs4 import BeautifulSoup
import pandas as pd
```

```python
os.listdir()[1:11]
```

```python


history=[]

for i in os.listdir()[1:11]:  
    with open(i, 'r') as f:   # Opening all 10 customer files
        xml_data=f.read()
        bs_data = BeautifulSoup(xml_data, 'xml') 
        payment_history=bs_data.find_all("COMBINED-PAYMENT-HISTORY")  # Getting Payment data From all XML files
        for i in payment_history:
                history.append(i.text)
                

History_split=[]
import re

# Removing Unwanted Object from data
for i in history:

    a=re.split(r'[,|,\s,'',]',i)
#     print(a,'\n')
    History_split.append(a)

        
for i in History_split:       # Removing unwanted using pop 
    removespace=i.pop(-1)
    

    
# Coverting Data Into dictionary (key:Value Format)
main_history=[]     
def Convert(lst):
    res_dct = map(lambda i: (lst[i], lst[i+1]), range(len(lst)-1)[::2])
    return dict(res_dct)
 
for i in History_split:
    sample=(Convert(i))
    main_history.append(sample)
    
    
data=pd.DataFrame(main_history)             # Creating DataFrame

# Replacing unwanted part of data into emptystring
for i in data.columns:           
    data[i]=data[i].str.replace('/','').astype(str)

# Replacing unwanted part of data into emptystring    
for i in data.columns:
    data[i]=(data[i].str.replace("[XXXSTDDDDSUBDBTLO]",""))
 

#removing nan string data into empty string
for i in data.columns:
    data[i]=(data[i].str.replace("[nan]",""))
    
    
list1=data.values.tolist()  # Converting list data into DataFrame with Pandas


NewList=[]                # Seperating Numeric Data For Calculation
for a in list1:
     for b in a:
          if b.isnumeric():
                NewList.append(b)
            
NewList=list(map(int,NewList))     # Converting Datatype String into integer
```

```python
df=pd.DataFrame(NewList) # CReating DataFrame
df.rename(columns={0: "Days_Past_Due"}, inplace=True) # Changing Column name in the DataFrame
```

```python
df       # Data of past dues
```

```python
# Calculating data which have  30+ DPD
trades_with_30Plus =0    
for i in NewList:
    if i >30:
        trades_with_30Plus+=1
trades_with_30Plus
```

```python
Answer=( trades_with_30Plus/len(NewList)*100)
print('trades are with 30+ DPD (more than 30 days past due) among all  the trades available',Answer,"%")
```

### Answer: Question 1
##  Percentage Trades are with 30+ DPD (more than 30 days past due) among all  the trades available = 15.143299767621999 %
